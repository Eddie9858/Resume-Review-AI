# 추론 코드

```python
server_code = r"""
import uvicorn
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from unsloth import FastLanguageModel
import torch
import nest_asyncio

nest_asyncio.apply()

app = FastAPI()

max_seq_length = 2048  # 선택 가능한 최대 시퀀스 길이
dtype = None  # None으로 자동 감지. Tesla T4, V100의 경우 Float16, Ampere+의 경우 Bfloat16
load_in_4bit = True  # 메모리 사용량을 줄이기 위해 4bit 양자화를 사용. False로 설정 가능.

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="/content/drive/MyDrive/model/checkpoint-500",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)

default_prompt1 = "당신은 유용한 AI 어시스턴트입니다. 사용자의 질의에 대해 친절하고 정확하게 답변해야 합니다. You are a helpful AI assistant, you'll need to answer users' queries in a friendly and accurate manner."
default_prompt2 = "너는 자기소개서 생성 AI이고 전체를 읽고 요구사항에 맞는 자기소개서를 이어서 생성해줘."

class Data(BaseModel):
    text: str

def chat1(prompt: str, text: str) -> str:
    alpaca_prompt = (
        "Below is an instruction that describes a task, paired with an input that provides further context. "
        "Write a response that appropriately completes the request.\n\n"
        "### Instruction:\n"
        "{instruction}\n\n"
        "### Input:\n"
        "{input}\n\n"
        "### Response:\n"
    )

    inputs = tokenizer(
        [
            alpaca_prompt.format(
                instruction=prompt,  # 지시사항
                input=text,          # 입력
            )
        ], return_tensors="pt").to("cuda")

    outputs = model.generate(**inputs, max_new_tokens=64, use_cache=True)
    decoded_output = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    response_index = decoded_output[0].find("### Response:")
    if response_index != -1:
        response = decoded_output[0][response_index + len("### Response:"):]
        return response.strip()
    return ""

def chat_multiple(text: str) -> str:
    # 첫 번째 추론
    context = "내가 만든 자기소개서야 보고 들어가면 좋을 것 같은 문장 작성해줘 " + text
    response = chat1(default_prompt1, context)
    responses = [response]

    # 두 번째 및 세 번째 추론
    context = response
    for _ in range(2):
        response = chat1(default_prompt2, context)
        responses.append(response)
        context = response

    # 결과를 하나의 문자열로 합침
    combined_response = "\n".join(responses)
    return combined_response

@app.post("/process")
async def process_data(data: Data):
    try:
        result = chat_multiple(data.text)
        return {"result": result}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

if __name__ == '__main__':
    uvicorn.run(app, host='0.0.0.0', port=8000, log_level="info")

"""

# server.py 파일로 저장
with open("server.py", "w") as f:
    f.write(server_code)

```

```python

# ngrok을 통해 서버를 외부에 노출시키기
from pyngrok import ngrok

# ngrok 인증 토큰 설정
NGROK_AUTH_TOKEN = os.getenv("NGROK_AUTH_TOKEN")
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# 기존 ngrok 터널 닫기
tunnels = ngrok.get_tunnels()
for tunnel in tunnels:
    ngrok.disconnect(tunnel.public_url)

# FastAPI 서버 실행 (백그라운드에서 실행)
get_ipython().system_raw('python server.py &')

# 새로운 ngrok 터널 열기
public_url = ngrok.connect(8000)
print(f"Public URL: {public_url}")

# ngrok 웹 인터페이스 보기 (옵션)
ngrok_tunnel = ngrok.get_tunnels()
print(ngrok_tunnel)
```
