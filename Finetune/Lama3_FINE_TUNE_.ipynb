# model,tokenizer 설정

```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "MLP-KTLim/llama-3-Korean-Bllossom-8B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
model = FastLanguageModel.get_peft_model(
     model,
     r = 16, # 아무 숫자나 선택할 수 있음! 추천값은 8, 16, 32, 64, 128
     target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                       "gate_proj", "up_proj", "down_proj",], # 타겟 모듈 지정
     lora_alpha = 16, # LoRA의 알파값
     lora_dropout = 0, # 드롭아웃 설정 (0으로 설정하면 최적화됨)
     bias = "none",    # 바이어스 설정 (최적화된 값은 "none")
     # [NEW] "unsloth"는 VRAM을 30% 적게 사용하며, 배치 크기를 2배 더 크게 설정할 수 있음!
     use_gradient_checkpointing = "unsloth", # 매우 긴 컨텍스트를 위해 True 또는 "unsloth" 설정
     random_state = 3407, # 랜덤 시드 값
     use_rslora = False,  # 순위 안정화된 LoRA를 지원
     loftq_config = None, # LoftQ 설정
 )
```

## 학습데이터 설정

```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token # EOS_TOKEN을 추가해야 함
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # EOS_TOKEN을 반드시 추가해야 함. 그렇지 않으면 생성이 무한히 계속될 수 있음!
        text = alpaca_prompt.format(instruction, input, output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
pass

#데이터셋 허깅페이스에 배포후 이용
from datasets import load_dataset
dataset = load_dataset("honggigi/kor-resume-4classification", split = "train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

## 학습 설정

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model, # 모델 설정
    tokenizer = tokenizer, # 토크나이저 설정
    train_dataset = dataset, # 학습 데이터셋 설정
    dataset_text_field = "text", # 데이터셋의 텍스트 필드 지정
    max_seq_length = max_seq_length, # 최대 시퀀스 길이 설정
    dataset_num_proc = 2, # 데이터셋 처리 프로세스 수
    packing = False, # 짧은 시퀀스에 대해 학습 속도를 5배 높일 수 있음
    args = TrainingArguments(
        per_device_train_batch_size = 64, # 각 장치당 배치 크기
        gradient_accumulation_steps = 16, # 그래디언트 누적 단계 수
        warmup_steps = 5, # 웜업 단계 수
        max_steps = 10, # 최대 학습 단계 수
        learning_rate = 2e-4, # 학습률
        fp16 = not torch.cuda.is_bf16_supported(), # fp16 설정 (bf16 미지원 시)
        bf16 = torch.cuda.is_bf16_supported(), # bf16 설정 (지원 시)
        logging_steps = 1, # 로그 기록 단계 수
        optim = "adamw_8bit", # 옵티마이저 설정
        weight_decay = 0.01, # 가중치 감소 설정
        lr_scheduler_type = "linear", # 학습률 스케줄러 타입 설정
        output_dir = "/content/drive/MyDrive/model/model4", # 출력 디렉토리 설정
    ),
)
```

## 학습

```python
trainer_stats = trainer.train()
```
